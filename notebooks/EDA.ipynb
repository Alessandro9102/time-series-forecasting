```python
import pandas as pd
df = pd.read_csv("../data/processed/master_merged.csv")

# Check Budapest
print("=== BUDAPEST ===")
print(df[df["nuts2"] == "HU11"].to_string())

# Check which countries have broadband missing
print("\n=== BROADBAND MISSING BY COUNTRY ===")
print(df[df["broadband_pct"].isna()]["country"].value_counts())

# Check which countries have it
print("\n=== BROADBAND PRESENT BY COUNTRY ===")
print(df[df["broadband_pct"].notna()]["country"].value_counts())

# Check all HU regions
print("\n=== ALL HUNGARIAN REGIONS ===")
print(df[df["country"] == "HU"].to_string())
```

```python
import pandas as pd
import eurostat

# Check what years broadband data exists for DE and PL
raw = eurostat.get_data_df("isoc_r_broad_h", flags=False, filter_pars={"unit": "PC_HH"})

# Find the geo column
geo_col = [c for c in raw.columns if "geo" in c.lower()][0]
raw = raw.rename(columns={geo_col: "nuts2"})
raw["nuts2"] = raw["nuts2"].astype(str).str.strip()

# Filter to NUTS2 only
raw = raw[raw["nuts2"].str.len() == 4]

# Check DE and PL coverage across years
for country in ["DE", "PL", "HR"]:
    subset = raw[raw["nuts2"].str.startswith(country)]
    year_cols = [c for c in raw.columns if str(c).isdigit() or (len(str(c))==4 and str(c).startswith("20"))]
    print(f"\n=== {country} broadband coverage by year ===")
    print(subset[["nuts2"] + year_cols].to_string())
```
