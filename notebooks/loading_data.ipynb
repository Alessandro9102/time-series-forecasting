```python
import eurostat
import pandas as pd

# 1. Download the dataset
print("Downloading isoc_ec_ib20...")
df = eurostat.get_data_df("isoc_ec_ib20")

# 2. Inspect structure
print("\n--- Shape ---")
print(df.shape)

print("\n--- Columns ---")
print(df.columns.tolist())

print("\n--- First rows ---")
print(df.head())

print("\n--- Unique values per dimension ---")
for col in df.columns:
    if df[col].dtype == object:
        print(f"\n{col}: {df[col].unique()[:10]}")  # first 10 unique values
```

```python
df_pandas = pd.DataFrame(df)
print("\n--- DataFrame info ---")   
df_pandas.info()
df_pandas.head(40)
```

```python
df_pandas.shape
df_pandas["unit"].unique()
```

```python
df_filtered = df_pandas[
    (df_pandas["unit"] == "PC_IND") &
    (df_pandas["indic_is"] == "I_BUY3") &
    (df_pandas["ind_type"] == "IND_TOTAL")
].copy()


print("\n--- Filtered DataFrame ---")
print(df_filtered.head())
print(df_filtered.shape)
```

```python
# Identify year columns
year_cols = [col for col in df_filtered.columns if str(col).startswith("20")]

# Reshape to long format
df_long = df_filtered.melt(
    id_vars=["geo\\TIME_PERIOD"],
    value_vars=year_cols,
    var_name="year",
    value_name="pct_online_buyers"
).rename(columns={"geo\\TIME_PERIOD": "country"})

df_long["year"] = df_long["year"].astype(int)
df_long = df_long.dropna(subset=["pct_online_buyers"])

print(df_long.sort_values(["country", "year"]))
```
