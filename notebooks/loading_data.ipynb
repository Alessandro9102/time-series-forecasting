```python
import eurostat
import pandas as pd

# 1. Obtener el índice completo de todos los datasets
print("Descargando índice de Eurostat...")
toc_df = eurostat.get_toc_df()
print(f"Total de datasets: {len(toc_df)}")
print(toc_df.head())

# 2. Buscar datasets relacionados con "online purchases"
print("\n--- Buscando 'online purchases' ---")
resultados = eurostat.subset_toc_df(toc_df, "online purchases")
print(resultados[['title', 'code']].head(10))

# 3. Buscar datasets relacionados con "e-commerce"
print("\n--- Buscando 'e-commerce' ---")
resultados_ecom = eurostat.subset_toc_df(toc_df, "e-commerce")
print(resultados_ecom[['title', 'code']].head(10))

# 4. Una vez tengas el código correcto, descargar datos
# Ejemplo con un dataset que SÍ funciona (GDP regional)
print("\n--- Descargando GDP regional (nama_10r_2gdp) ---")
df_gdp = eurostat.get_data_df("nama_10r_2gdp")
print(df_gdp.head())
```
